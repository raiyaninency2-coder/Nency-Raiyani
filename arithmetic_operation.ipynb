```python
import numpy as np
```

# arithmetic operation

```python
var = np.array([1,2,3,4])
varadd = var + 3
print(varadd)
```

```python
var1 = np.array([98,97])
var2 = np.array([92,88])
total = var1 + var2
minus = var1 - var2

print(total)
print(minus)
```

```python
a = np.add(var1,var2)
b = np.subtract(var1,6)
c = np.multiply(var1,var2)
d = np.divide(var1,var2)
e = np.mod(var1,var2)
f = np.power(var1 , 2)
g = np.reciprocal(var1)


print(a,b,c,d,e,f,g)
```

# 2D array

```python
arr1 = np.array([[1,2,3,4] , [5,6,7,8]])
arr2 = np.array([[1,2,3,4] , [5,6,7,8]])
t = arr1 + arr2
print(t)
```

# other function

```python
arr3 = np.array([11,2,3,78,89,86])
print(np.max(arr3))
print(np.min(arr3))
print(np.argmax(arr3))
print(np.argmin(arr3))
print(np.sqrt(arr3))



print(np.sin(arr3))


print(np.cumsum(arr3))

```
