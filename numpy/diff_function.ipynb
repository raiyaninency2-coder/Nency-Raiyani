```python
import numpy as np
```

# zeros

```python
ar_z = np.zeros(4)
ar_z1 = np.zeros((4,4))
print(ar_z)
print(ar_z1)
```

# all elements be one in matrix

```python
ar_1 = np.ones((3,4))
print(ar_1) 
```

# empty array 

```python
ar_2 = np.empty(5)
print(ar_2)
```

# range function

```python
ar_3 = np.arange(1,4,2)
ar_3
```

# diagonal array using   {eye} function

```python
ar_4 = np.eye(4)
ar_5 = np.eye(3,5)
ar_4
ar_5
```

# linespace that use in n part of value range

```python
ar_6 = np.linspace(0,20,num=10)
ar_6
```
