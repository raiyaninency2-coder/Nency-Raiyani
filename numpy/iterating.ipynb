# iterating

```python
import numpy as np
```

```python
# for one dimension
var = np.array([1,4,9,16,25])
for i in var:
    print(i)
```

```python
# for two dimension
var1 = np.array([[1,2] , [3,4]])
for j in var1:
    print()
    for k in j:
        print(k)
```

```python
var2 = np.array([[[1,4] , [1,8] , [1,16] , [1,32]]])
for i in var2:
    print()
    for j in i:
        print()
        for k in j:
            print(k)
```

```python
# using function   nditer
var3 = np.array([[[1,4] , [1,8] , [1,16] , [1,32]]])
for i in np.nditer(var3):
    print(i)

```

```python
# iterating with indexing  using function ndenumerate
var4 = np.array([[[1,4] , [1,8] , [1,16] , [1,32]]])
for i,d in np.ndenumerate(var4):
    print(i,d)
```
