# copy vs new

```python
import numpy as np
```

```python
# copy
# if we change in var , co has not change.
var = np.array([1,4,8,16,32])
co = np.copy(var)
var[1]= 98
print("var : " , var)
print("copy : " ,co)
```

```python
# view
#if we change in var1 , vi automatically change
var1 = np.array([1,4,8,16,32])
vi = var1.view
print("var : " , var)
print("view : " ,vi)
```
