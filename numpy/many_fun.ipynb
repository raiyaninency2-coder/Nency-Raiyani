```python
import numpy as np
```

```python
# search   :  using where function , in result index are returned
var = np.array([1,2,3,2,4,5,2,6,2,7])
x = np.where(var == 2)
y = np.where(var%2 == 0)
print(x)
print(y)
```

```python
# choose place for sorted array to being sorted array
var1 = np.array([1,2,4,6,8,10])
p = np.searchsorted(var1,3)
q = np.searchsorted(var1,9)
print(p , q)
```

# sort

```python
var2 = np.array([92,97,98,37.75,35.25,36.50])
print(np.sort(var2))


# also for string:
var3 = np.array(["nency","niyati","vraj","ayaan","dev"])
print(np.sort(var3))
```

```python
# sorting for two dimentional array
var4 = np.array([[92,97],[98,37.75],[35.25,36.50]])
print(np.sort(var4,axis=0))
print()
print(np.sort(var4,axis=1))
```

# filter function

```python
var5 = np.array(["nency" , "dev"])
f = [True , False]
s = var5[f]
print(s)
```
