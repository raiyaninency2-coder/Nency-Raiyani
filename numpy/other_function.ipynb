```python
import numpy as np
```

# shuffle

```python
# shuffling the all elements in rendomly
var = np.array([1,2,3,4,5])
np.random.shuffle(var)
print(var)
```

# unique

```python
var1 = np.array([1,2,3,4,2,5,7,1,3,7,8,2])
x = np.unique(var1,return_index=True,return_counts=True)
print(x)
```

# Resize

```python
var2 = np.array([1,2,3,4,5,6])
y = np.resize(var2,(3,2))
z = np.resize(var2,(6,1))
print(y)
print()
print(z)
```

# flatten and ravel

```python
# flatten convert two dimensional array to one dimensional array
# y and z is two dimensional array
print(y.flatten())
print(z.flatten())


print(y.ravel())
```
