# split

```python
import numpy as np
```

```python
var = np.array([1,2,3,4,5,6,7])
print(var)
print()
ar = np.array_split(var,2)
ar1 = np.array_split(var,4)
ar2 = np.array_split(var,5)
print(ar)
print(ar1)
print(ar2)
print(ar[0])
print(ar1[1])
```

```python
var1 = np.array([[1,2],[3,4],[5,6]])
print(var1)
print()
ar3 = np.array_split(var1,2)
ar4 = np.array_split(var1,2,axis = 1)
print(ar3)
print(ar4)
```
